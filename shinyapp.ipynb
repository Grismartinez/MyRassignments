```r
# Load required libraries
library(shiny)
library(shinydashboard)


# UI (User Interface)
ui <- dashboardPage(
  dashboardHeader(title = "Iris Data Dashboard"),
  
  dashboardSidebar(
    sidebarMenu(
      menuItem("Dashboard", tabName = "dashboard", icon = icon("dashboard")),
      
      # Interactive filter
      selectInput("species", 
                  "Select Species:", 
                  choices = c("All", unique(iris$Species)),
                  selected = "All")
    )
  ),
  
  dashboardBody(
    tabItems(
      tabItem(tabName = "dashboard",
              
              fluidRow(
                box(title = "Sepal Length Distribution", status = "primary", solidHeader = TRUE,
                    plotOutput("histPlot")),
                
                box(title = "Sepal vs Petal Length", status = "primary", solidHeader = TRUE,
                    plotOutput("scatterPlot"))
              ),
              
              fluidRow(
                box(title = "Average Measurements by Species", status = "primary", solidHeader = TRUE,
                    plotOutput("barPlot"))
              )
      )
    )
  )
)

# Server logic
server <- function(input, output) {
  
  # Reactive data filtering
  filtered_data <- reactive({
    if (input$species == "All") {
      iris
    } else {
      iris %>% filter(Species == input$species)
    }
  })
  
  # Histogram
  output$histPlot <- renderPlot({
    ggplot(filtered_data(), aes(x = Sepal.Length, fill = Species)) +
      geom_histogram(bins = 20, alpha = 0.7) +
      theme_minimal() +
      labs(title = "Distribution of Sepal Length")
  })
  
  # Scatter Plot
  output$scatterPlot <- renderPlot({
    ggplot(filtered_data(), aes(x = Sepal.Length, y = Petal.Length, color = Species)) +
      geom_point(size = 3) +
      theme_minimal() +
      labs(title = "Sepal Length vs Petal Length")
  })
  
  # Bar Plot (average values)
  output$barPlot <- renderPlot({
    avg_data <- filtered_data() %>%
      group_by(Species) %>%
      summarise(across(where(is.numeric), mean))
    
    ggplot(avg_data, aes(x = Species, y = Sepal.Length, fill = Species)) +
      geom_bar(stat = "identity") +
      theme_minimal() +
      labs(title = "Average Sepal Length by Species")
  })
}
shinyApp(ui = ui, server = server)
```
